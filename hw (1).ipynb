```python
aapl_df = aapl_df.rename(columns={"Open": "Open_AAPL", "High": "High_AAPL", "Low": "Low_AAPL", "Close": "Close_AAPL", "Adj Close": "Adj_Close_AAPL", "Volume": "Volume_AAPL", "Dividends": "Dividends_AAPL", "Stock Splits": "Stock_Splits_AAPL"})
btc_df = btc_df.rename(columns={"Open": "Open_BTC", "High": "High_BTC", "Low": "Low_BTC", "Close": "Close_BTC", "Volume": "Volume_BTC"})
gas_df = gas_df.rename(columns={"Date": "date", "Open": "Open_GAS", "High": "High_GAS", "Low": "Low_GAS", "Close": "Close_GAS", "Adj Close": "Adj_Close_GAS", "Volume": "Volume_GAS"})
```

```python
df = aapl_df.merge(btc_df, on = 'date', how = 'outer')
df = df.merge(gas_df, on = 'date', how = 'outer')
df.tail()
```

```python
df.isna().sum()
```

```python
df.fillna(df.mean(), inplace=True)
df.isna().sum()
```

```python
df['btc_profit'] = np.log(df['Close_BTC']) - np.log(df['Open_BTC'])
df['btc_profit'].describe()
```

Таким образом, мю = 0.000640, а сигма = 0.038682.

№2

```python
from scipy.stats import shapiro
def shapiro_test(spisok):
  stat, p = shapiro(spisok)
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Normal distribution')
  else:
    print('Not a normal distribution')

shapiro_test(df['btc_profit'].tolist())
```

```python
# Приводим данные к стандартному распределению
df['btc_profit_standart'] = (df['btc_profit'] - df['btc_profit'].mean()) / df['btc_profit'].std()
```

```python
# Находим межквартильный размах, а также левые и правые границы(крайние границы, такие что значения, выходящие за них, будут считаться выбросами)
q_25 = df['btc_profit_standart'].quantile(0.25)
q_75 = df['btc_profit_standart'].quantile(0.75)
interquartile_range = q_75 - q_25
left_border = q_25 - 1.5*interquartile_range
right_border = q_75 + 1.5*interquartile_range
```

```python
# Составим новый датасет, в который будем записывать все значения кроме выбросов.
df_edited = df.copy()
for i in range(len(df)):
  val = df.loc[i, 'btc_profit_standart']
  if not ((val > left_border) and (val < right_border)):
    df_edited = df_edited.loc[df_edited['btc_profit_standart'] != val]
```

```python
# Попробуем нормализовать данные, используя дробное дифференцирование со степенью дифференцирования n = 1.
from fracdiff import fdiff

shapiro_test(fdiff(np.array(df_edited['btc_profit_standart'].tolist()), n = 1))
```

```python
df_edited['aaple_profit'] = np.log(df_edited['Close_AAPL']) - np.log(df_edited['Open_AAPL'])
df_edited['gas_profit'] = np.log(df_edited['Close_GAS']) - np.log(df_edited['Open_GAS'])
df_edited['aaple_profit_standart'] = (df_edited['aaple_profit'] - df_edited['aaple_profit'].mean()) / df_edited['aaple_profit'].std()
df_edited['gas_profit_standart'] = (df_edited['gas_profit'] - df_edited['gas_profit'].mean()) / df_edited['gas_profit'].std()
```

```python
btc_fdiff = fdiff(np.array(df_edited['btc_profit_standart'].tolist()), n = 1)
aaple_fdiff = fdiff(np.array(df_edited['aaple_profit_standart'].tolist()), n = 1)
gas_fdiff = fdiff(np.array(df_edited['gas_profit_standart'].tolist()), n = 1)
```

```python
d = {'btc_fdiff' : btc_fdiff, 'aaple_fdiff' : aaple_fdiff, 'gas_fdiff' : gas_fdiff}
fdiff_df = pd.DataFrame(data=d)
```

```python
from scipy import stats
# Проверка гипотезы о равентсве средних
print(stats.f_oneway(fdiff_df['aaple_fdiff'], fdiff_df['btc_fdiff'], fdiff_df['gas_fdiff']))
```

p-value не менее 0.05, значит, что мы не можем отвергнуть нулевую гипотезу. Примем, что средние трех доходностей равны.

```python
# Проверка гипотезы о равентсве дисперсий
print(stats.levene(fdiff_df['aaple_fdiff'], fdiff_df['btc_fdiff'], fdiff_df['gas_fdiff']))
```
